```python
import pickle
import numpy as np
import pandas as pd
from fastapi import FastAPI
from pydantic import BaseModel
import uvicorn
import nest_asyncio

# Enable running FastAPI inside Jupyter Notebook
nest_asyncio.apply()

# Load trained model
with open("xgb_model.pkl", "rb") as f:
    model = pickle.load(f)

# Initialize FastAPI app
app = FastAPI()

# Define input structure
class TripRequest(BaseModel):
    pickup_latitude: float
    pickup_longitude: float
    dropoff_latitude: float
    dropoff_longitude: float
    passenger_count: int
    hour: int
    day_of_week: int
    month: int
    is_weekend: int
    haversine_distance: float
    vendor_1: int
    vendor_2: int

@app.post("/predict")
def predict_trip_duration(data: TripRequest):
    # Convert input data to model format
    features = np.array([[
        data.pickup_latitude, data.pickup_longitude, data.dropoff_latitude,
        data.dropoff_longitude, data.passenger_count, data.hour, data.day_of_week,
        data.month, data.is_weekend, data.haversine_distance, data.vendor_1, data.vendor_2
    ]])
    
    # Predict trip duration
    predicted_duration = model.predict(features)[0]
    
    return {"predicted_trip_duration": float(predicted_duration)}

# Run FastAPI server inside Jupyter Notebook
uvicorn.run(app, host="0.0.0.0", port=8000)
```
